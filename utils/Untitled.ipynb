```python
# quick rename

import os

def rename_files(input_dir):
    """
    Renames files in the specified directory by removing 'texts' from their filenames.
    
    Args:
    input_dir (str): The directory containing files to rename.
    """
    # Loop through all files in the directory
    for filename in os.listdir(input_dir):
        if 'channel,' in filename:
            # Construct the new filename
            new_filename = filename.replace('channel,', '')
            old_file_path = os.path.join(input_dir, filename)
            new_file_path = os.path.join(input_dir, new_filename)
            
            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f"Renamed '{filename}' to '{new_filename}'")

# Example usage
input_dir = r'D:\Chang_files\workspace\data\InSillico\Rubin\test\Brightfield_Patch'
rename_files(input_dir)
```
