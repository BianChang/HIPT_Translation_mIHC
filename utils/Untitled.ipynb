```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Path to the image
image_path = r'F:\projects\In silico project\data\train\Rubin\scott_1_0\lab-Rubin,condition-scott_1_0,acquisition_date,year-2016,month-2,day-6,well-r02c05,depth_computation,value-MAXPROJECT,is_mask-false,kind,value-ORIGINAL.png'

# Load the image
img = Image.open(image_path)

# Display the image
plt.imshow(img)
plt.title('Original Image')
plt.show()

# Image information
print(f"Image size (Width x Height): {img.size}")
print(f"Number of Channels: {'Grayscale' if img.mode == 'L' else '3 (RGB)' if img.mode == 'RGB' else img.mode}")

# Display each channel separately if it's an RGB image
if img.mode == 'RGB':
    img_np = np.array(img)
    for i, color in enumerate(['Red', 'Green', 'Blue']):
        channel_img = np.zeros(img_np.shape, dtype='uint8')
        channel_img[:,:,i] = img_np[:,:,i]
        plt.imshow(channel_img)
        plt.title(f'{color} Channel')
        plt.show()
```
